```python
categorical_columns=["Sex", "Housing", "Saving accounts", "Checking account", "Purpose"]
```

```python
features=["Age", "Sex", "Job", "Housing", "Saving accounts", "Checking account",
       "Credit amount", "Duration", "Purpose"]
target="Risk"
print(features)
print(target)
```

```python
df_model=df[features+[target]].copy()
df_model.head()
```

```python
categorical_columns=df_model.select_dtypes(include="object").columns.drop("Risk")
```

```python
!pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
for col in categorical_columns:
    df_model[col]=encoder.fit_transform(df_model[col].astype(str))
df_model[col]
```

```python
df_model[target].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x=df_model.drop(target,axis=1)
y=df_model[target]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
model=LogisticRegression()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
def train_model(model,param_grid,x_train,x_test,y_train,y_test):
    grid= GridSearchCV(model,param_grid,cv=5,scoring="accuracy",n_jobs=-1)
    grid.fit(x_train,y_train)
    best_model=grid.best_estimator_
    y_pred=best_model.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    class_report=classification_report(y_test,y_pred)
    con_matrix=confusion_matrix(y_test,y_pred)
    
    return best_model,acc,class_report,con_matrix,grid.best_params_
    
```

```python
dt= DecisionTreeClassifier(random_state=1,class_weight="balanced")
dt_param_grid={
    "max_depth":[3,5,7,10,None],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}
```

```python
best_dt,acc_dt,class_report_dt,con_matrix_dt,params_dt=train_model(dt,dt_param_grid,x_train,x_test,y_train,y_test)
```

```python
print("accuracy_score",acc_dt)
print("class_report",class_report_dt)
print("con_matrix",con_matrix_dt)
print("best_parameters:",params_dt)
```

```python
rf=RandomForestClassifier(random_state=1,class_weight="balanced",n_jobs=-1)
rf_param_grid={
    "n_estimators":[50,100],
    "max_depth":[10,None],
    "min_samples_split":[2,5],
    "min_samples_leaf":[1,2]
}
```

```python
best_rf,acc_rf,class_report_rf,con_matrix_rf,params_rf=train_model(rf,rf_param_grid,x_train,x_test,y_train,y_test)
```

```python
print("accuracy_score",acc_rf)
print("class_report",class_report_rf)
print("con_matrix",con_matrix_rf)
print("best_parameters:",params_rf)
```

```python
et=ExtraTreesClassifier(random_state=1,class_weight="balanced",n_jobs=-1)
et_param_grid={
    "n_estimators":[100],
    "max_depth":[10,None],
    "min_samples_split":[2,5],
    "min_samples_leaf":[1,2]
}
```

```python
best_et,acc_et,class_report_et,con_matrix_et,params_et=train_model(et,et_param_grid,x_train,x_test,y_train,y_test)
```

```python
print("accuracy_score",acc_et)
print("class_report",class_report_et)
cm=con_matrix_et
print("con_matrix",cm)
print("best_parameters:",params_et)
```
